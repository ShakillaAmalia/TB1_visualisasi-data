```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(r'C:\Users\Kyla\iris\iris.data', header=None, names=column_names)

stats = df.describe().loc[['min', 'max', 'mean', 'std']]

stats.loc['min'] = stats.loc['min'].round(1)
stats.loc['max'] = stats.loc['max'].round(1)  
stats.loc['mean'] = stats.loc['mean'].round(2)  
stats.loc['std'] = stats.loc['std'].round(2)  
print("Statistik Dasar:\n", stats)

correlations = {}
df['species_code'] = pd.Categorical(df['species']).codes
for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    correlation, _ = pearsonr(df[col], df['species_code'])
    correlations[col] = round(correlation, 4)

print("\nKorelasi dengan Kelas:")
print(correlations)

stats.loc['correlation'] = correlations.values()
stats
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'C:\Users\Kyla\iris\iris.data'
column_names = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Class"]
iris_data = pd.read_csv(file_path, header=None, names=column_names)

sepal_length_means = iris_data.groupby("Class")["Sepal Length"].mean().reset_index()

sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(data=sepal_length_means, x='Class', y='Sepal Length', palette="viridis")
plt.title("Bar Chart")
plt.xlabel("Iris Species")
plt.ylabel("Sepal Length")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'C:\Users\Kyla\iris\iris.data'
column_names = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Class"]
iris_data = pd.read_csv(file_path, header=None, names=column_names)

plt.figure(figsize=(8, 5))
plt.hist(iris_data['Sepal Length'], bins=10, color='blue', edgecolor='black')
plt.title("Sepal Length")
plt.xlabel("Points")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

for column in column_names[:-1]: 
    plt.plot(iris_data.index, iris_data[column], label=column)

plt.title("Iris Dataset Fesature")
plt.xlabel
plt.ylabel("Value (cm)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

class_colors = {"Iris-setosa": "red", "Iris-versicolor": "green", "Iris-virginica": "blue"}
sns.scatterplot(data=iris_data, x="Sepal Length", y="Sepal Width", hue="Class", palette=class_colors)
plt.title
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")
plt.show()
```
